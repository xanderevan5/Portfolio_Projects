```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots

df = pd.read_csv(r'C:\Users\evan6\Downloads\movies.csv')
```

```python
print(df.columns)
df.head()
```

```python
# Let's see if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Change budget to int
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Create correct YYYY year column
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Budget high correlation
# Company high correlation
```

```python
# Scatter plot budget vs. gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')


plt.show()
```

```python
#Plot budget vs. gross using seaborn

sns.regplot(x='budget',
            y='gross',
            data=df,
            scatter_kws={'color': 'red'},
            line_kws={'color': 'blue'})
```

```python
df.corr(method= 'pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross
#  	0.740247
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix between Numeric Movie Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```
